```python
import pandas as pd 
import numpy as np
```

```python
dftrain = pd.read_csv("/content/train.csv")
dftest = pd.read_csv("/content/test.csv")
print(dftrain.isnull().sum())
print(dftest.isnull().sum())
dftrain.drop("Cabin", axis = 1, inplace = True)
dftest.drop("Cabin", axis = 1, inplace = True)
dftrain["Embarked"].fillna("S", inplace = True)
!pip3 install impyute
dftrain.head(10)
```

```python
dftest.head(12)
```

```python
import sys
from impyute.imputation.cs import fast_knn
sys.setrecursionlimit(100000) #Increase the recursion limit of the OS

# start the KNN training
knntrain=fast_knn(dftrain[["PassengerId","Age"]].values, k=30)
knntest=fast_knn(dftest[["PassengerId","Age"]].values, k=30)
knntrain = knntrain.astype(int)
knntrain = knntrain.astype(float)
knntest = knntest.astype(int)
knntest = knntest.astype(float)
```

```python
dftrain['Age'] = knntrain[:,1]
dftest["Age"] = knntest[:,1]
dftrain.head(10)
```

```python
dftest.head(12)
```

```python
dftest["Fare"].fillna(np.mean(dftest["Fare"]), inplace = True)
print(dftrain.isna().sum())
print(dftest.isna().sum())
```

```python
dftrain.drop("Name", axis = 1, inplace = True)
dftrain.drop("Ticket", axis = 1, inplace = True)
dftest.drop("Name", axis = 1, inplace = True)
dftest.drop("Ticket", axis = 1, inplace = True)
dftrain.drop("PassengerId", axis = 1, inplace = True)
yPassId = dftest["PassengerId"]
dftest.drop("PassengerId", axis = 1, inplace = True)
```

```python
dftrain.head(10)
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
dftrain[["Sex"]] = encoder.fit_transform(dftrain[["Sex"]])
dftest[["Sex"]] = encoder.fit_transform(dftest[["Sex"]])
dftrain[["Embarked"]] = encoder.fit_transform(dftrain[["Embarked"]])
dftest[["Embarked"]] = encoder.fit_transform(dftest[["Embarked"]])
dftrain[["Pclass"]] = encoder.fit_transform(dftrain[["Pclass"]])
dftest[["Pclass"]] = encoder.fit_transform(dftest[["Pclass"]])
dftrain[["SibSp"]] = encoder.fit_transform(dftrain[["SibSp"]])
dftest[["SibSp"]] = encoder.fit_transform(dftest[["SibSp"]])
dftrain[["Parch"]] = encoder.fit_transform(dftrain[["Parch"]])
dftest[["Parch"]] = encoder.fit_transform(dftest[["Parch"]])
```

```python
dftrain.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

X = dftrain.iloc[:,1:].values
y = dftrain.iloc[:,0].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
Xf = dftest.iloc[:][:].values
Yf = classifier.predict(Xf)
```

```python
dffinal = pd.DataFrame()
dffinal["PassengerId"] = yPassId
dffinal["Survived"] = Yf
dffinal
```

```python
dffinal.to_csv("submission.csv", index = False)
```
